## Estimator + Predictor APIs

(Almost?) All the supervised learning models (and some others) in the `scikit-learn` module implements the simple *Estimator API* and the *Predictor API*. These give us a consistent ways of using these models. It is very easy to use:

1. Pick the model that implements the `estimator` that suits your problem and `import` it.
2. Select the hyperparamaters of the model (we will later pick these automatically as well)
3. Prepare the data 
4. Call the `fit` function with the data
5. Use the model for inference with the `predict` (or `predict_proba`) function. 

(Note1: Transformer API is similar)  
(Note2: We skipped some steps, what are they?)

Let's follow these steps on a linear regression problem

Yet again the accuracy scores and individual predictions look good for the training data. However, we want the performance of the model on unseen data. To do this, we split the data into training set and testing set (and also validation set, which will come in a bit). 

`scikit-learn` has the functionality through the `train_test_split` class. The `test_size` parameter is used to set the relative test set size:

```python
#use cross_validation if model_selection does not work
from sklearn.model_selection import train_test_split

# The split is random which is what we want. 
# However, sometimes we need the same split for debugging purposes
# We can do this by setting the random seed (or random state per scikit-learn)
# By the way, we usually do not set the random seed

# Let's do a 80%-20% split
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, 
                                                test_size = 0.2, # relative size of the test set wrt the entire set
                                                random_state = 12) # random seed

# Train
logregModel.fit(Xtrain, ytrain)

#Performance evaluation with the test set
ypred = logregModel.predict(Xtest)
print(accuracy_score(ytest, ypred))
print()
print(ytest)
print(ypred)
```

This also looks good! Is it too good to be true? Could this be due to the specific split? Let's try more splits to be sure

```python
# With another random seed 
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size = 0.2, random_state = 123)
logregModel.fit(Xtrain, ytrain)

ypred = logregModel.predict(Xtest)
print(accuracy_score(ytest, ypred))
print()
print(ytest)
print(ypred)
```

Let's get back to the random split case. We wanted multiple splits to be able trust our model more. Ideally, we want to get a rich diversity of data points for training and testing when we are evaluating the performance of our learned models. This is called cross-validation. 

There are multiple ways of splitting the data for cross-validation. We tried 10 random splits above. This is sometimes called shuffle splitting. Since we were working on toy data, this was not an issue. However, larger datasets and more complicated models would take a while to learn, limiting our options for random splits.

K-Fold Cross-validation is another, perhaps more commonly used, alternative. It is a systematic way to trade-off training and test set diversity while keeping the number of splits relative low. The idea is to divide the data into k equal-sized parts, keep one part as the test set and the remaining parts as the training set. For example when K = 5:



There are other cross-validation methods and other options for these as well. For example we can use stratification for each of these. Luckily for us `scikit-learn` has these on top of other cross-validation functionality. Let's start with  `cross_val_score`.

Note: Cross-validation is also useful for hyper-parameter selection as we will see

```python
from sklearn.model_selection import cross_val_score
crossValScores = cross_val_score(logregModel, X, y, cv = 5)
print("Scores: ", crossValScores)
print("Mean and 2 std: %0.2f (+/- %0.2f)" 
      % (crossValScores.mean(), crossValScores.std() * 2))
```

When we set the `cv = 5`, we told the method that we wanted to do 5-fold CV. We can also use CV classes for more control. For example there is the `ShuffleSplit` which splits the data with the given ratio with the given number of times. There is also `KFold` that does K-Fold cross validations. 

Let's see these in action.

```python
from sklearn.model_selection import ShuffleSplit, KFold

# Reset the random seed
import time
np.random.seed(int(time.time()))

# Random Splitting 75-25 5 times:
cv = ShuffleSplit(n_splits = 5, test_size = 0.25)
print("Shuffle Split: ")
print(cross_val_score(logregModel, X, y, cv=cv))
print()

# Random shuffling than dividing into 5 folds 
# Pay attention to shuffle = True!!!
cv = KFold(n_splits = 5, shuffle = True)
print("Shuffled K-Fold Split: ")
print(cross_val_score(logregModel, X, y, cv = cv))
```

These do not use stratification! 

There are stratified versions as well:

```python
from sklearn.model_selection import StratifiedShuffleSplit, StratifiedKFold

cv = StratifiedShuffleSplit(n_splits = 5, test_size = 0.25)
print("Stratified Shuffle Split: ")
print(cross_val_score(logregModel, X, y, cv = cv))

cv = StratifiedKFold(n_splits = 5, shuffle = True)
print("Stratified K-Fold Split: ")
print(cross_val_score(logregModel, X, y, cv = cv))
```

When we have low number of data points, we want to utilize as much data as possible for training. However, this adversely effects our ability to do testing. Leave-one-out cross validation is used in these cases where we keep only a single data point as test data and rotate each data point (equivalent N-Fold CV where N is the number of data points)

```python
from sklearn.model_selection import LeaveOneOut
cv = LeaveOneOut()
scores = cross_val_score(logregModel, X, y, cv = cv)
print(scores.mean())
```

##  Hyperparameter Selection

The performance tests so far were for static hyperparameters. However, hyperparameter selection is an important aspect in machine learning! (Why?)

 `scikit-learn` offers functionality for hyperparameter selection. Two most common methods:
1. Grid Search: Trying all combinations of given hyperparameter values and finding the best combination according to the given performance metric. The relevant class is `GridSearchCV`
2. Random Search: Sampling from given hyperparameter distributions given number of times and finding the best combination according to the given performance metric. `RandomizedSearchCV`

Let's start with GridSearch. The main hyperparameter of `LogisticRegression` is the regularization strength. The hyperparameter denoted with `C` by the class is inversely related to regularization (Low `C` -> stronger regularization). We are going to search for this. Just to show we can search for multiple hyperparameters at the same time, we are also going to set the solver as one of the hyperparameters.

Note: We are going to assume that the data is only the training data.

```python
from sklearn.model_selection import GridSearchCV

np.random.seed(int(time.time()))

# The performance test we pick 
cv = StratifiedKFold(n_splits = 5, shuffle = True)

paramsToTest = {"C": [0.1, 0.5, 1.0, 2.5, 5.0],
                "solver": ["sag", "lbfgs"]}

gscv = GridSearchCV(LogisticRegression(multi_class = "auto",  
                                       max_iter = 10000), 
                    paramsToTest, cv = cv, return_train_score = True)

gscv.fit(X, y)

print(gscv.best_params_)
print("Best C:", gscv.best_estimator_.C)
print("Best solver:", gscv.best_estimator_.solver)
```

Once we find the best hyperparameter set, we should use the entirety of the training data retrain with this set. This can be done automatically by setting the `refit` field of the hyper parameter selection classes (default = `True` so refitting is done by default).

Detailed Hyperparemeter selection statistics can be accessed over the `gscv.cv_results_` data structure.

```python
sorted(gscv.cv_results_.keys())
```

Let's take a look at a subset

```python
print(gscv.cv_results_['params'])
print()
print(gscv.cv_results_['mean_test_score'])
print()
print(gscv.cv_results_['std_test_score'])
```

```python
print(gscv.cv_results_['split1_train_score'])
print(gscv.cv_results_['split1_test_score'])
```

We can plot validation curves and visualize the results with this information. `scikit-learn` offers functionality for this via `validation_curve`. We will use this when we get to regression.

For random search, we need to define distributions. Values for each hyperparameter, that a distribution is defined for, is sampled randomly from these distributions for a single iteration

One way to do it is to provide a list of values. During the search, a value is selected uniformly randomly from this list. We can also provide an explicit distribution. During the search, a value is sampled from this distribution. 

The distributions must provide a sample using the `rvs` methods. The distributions under the `scipy.stats` submodule provides these. We can also implement our own sampling class and implement a `rvs` method. 

Additionally we specify the number of iterations through the `n_iter` argument.

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform as sp_uniform 

np.random.seed(int(time.time()))

cv = StratifiedKFold(n_splits = 5, shuffle = True)

paramsToTest = {"C": np.arange(0.1, 5.1, 0.1),
                "solver": ["sag", "lbfgs"]}

rscv = RandomizedSearchCV(LogisticRegression(multi_class = "auto", 
                                             max_iter = 10000), 
                          paramsToTest, cv = cv, n_iter = 10)

rscv.fit(X, y)

print("Best C with list: ", rscv.best_estimator_.C)
print("Best solver:", rscv.best_estimator_.solver)
```

```python
paramsToTest = {"C": sp_uniform(0.1, 4.9),
                "solver": ["sag", "lbfgs"]}

rscv = RandomizedSearchCV(LogisticRegression(multi_class = "auto", 
                                             max_iter = 10000), 
                          paramsToTest, cv = cv, n_iter = 10)
rscv.fit(X,y)
print("Best C with distribution: ", rscv.best_estimator_.C)
print("Best solver:", rscv.best_estimator_.solver)
```

So far we have not supplied the metric we want to use to pick the hyperparameters. We can do it with the `scoring` argument. If we do not supply anything, the `score` function of the estimator is used. It is the accuracy score for the Logistic Regression model.

## Model Evaluation

So far we have seen performance evaluation and hyper-parameter selection. Let's combine them by using a figure from `scikit-learn` resources,.



Furthermore, you can perfrom this multiple times with different initial train-test splits.

On the other hand, all of these steps may not be suitable for complex models and/or a large amount of data due to time concerns.

Let's apply this to iris dataset with the `LogisticRegression` model by:
* Split the data 60%-40% in a stratified manner
* Pick the regularization parameter C by using RandomSearch (Use uniform distribution between 0.1 and 5 as your distribution) and 3-Fold Cross validation
* Re-train the model with the picked hyperparameter using all the training data
* Use the test set to measure the performance of this model

```python
# What we need if we were to write it from scratch
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, auc
from sklearn.model_selection import train_test_split, StratifiedKFold, RandomizedSearchCV
from scipy.stats import uniform as sp_uniform

# Load the data
iris_data = datasets.load_iris()
y = iris_data["target"]
X = iris_data["data"]

# Split the data while preserving the target distribution
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size = 0.4, stratify = y)

# Set the cross validation method
cv = StratifiedKFold(n_splits = 3, shuffle = True)
 
# Setup the hyperparameter search over the training set
paramsToTest = {"C": sp_uniform(0.1, 4.9)}
rscv = RandomizedSearchCV(LogisticRegression(multi_class = "auto", solver = "lbfgs", max_iter = 10000),
                          paramsToTest, 
                          cv = cv, 
                          n_iter = 10, 
                          scoring = "roc_auc_ovr") #multi-class o-v-r

# Start the search and select
rscv.fit(Xtrain, ytrain)
print("Picked C: ", rscv.best_estimator_.C)
```

```python
print(rscv.cv_results_["mean_test_score"])
```

```python
# The rscv.best_estimator_ is trained with the entire input data, if we want to access it
# However, we can call the predict function over rscv object directly
# For example to test the performance:
ypred = rscv.predict(Xtest)
print(accuracy_score(ytest, ypred))
print(rscv.score(Xtest,ytest)) # compare with print(rscv.score(Xtrain,ytrain))
```

```python
# Let's look at the confusion matrix
from sklearn.metrics import confusion_matrix

mat = confusion_matrix(ytest, ypred)

plt.figure(figsize = (12, 8))
sns.heatmap(mat, square = True, annot = True,
            fmt = "d", cbar = False,
            xticklabels = iris_data["target_names"],
            yticklabels = iris_data["target_names"])
plt.xlabel("Predicted Class")
plt.ylabel("True Class")
plt.show()
```
